````python
import numpy as np
import keras

(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()
````

````python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(1, 9, figsize = (9, 1))
axs = axs.flatten()

for i in range(9):
    plt.sca(axs[i])
    plt.imshow(x_train[i], cmap = 'binary')
    plt.axis('off')
plt.show()
````

````python
x_train[0]
````

````python
x_train = x_train / 255
x_test = x_test / 255
````

````python
x_train = x_train.reshape(x_train.shape[0], -1)
x_test = x_test.reshape(x_test.shape[0], -1)
````

````python
import time

import xgboost as xgb
import lightgbm as lgb
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import ExtraTreeClassifier
from sklearn.svm import LinearSVC
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier

xgb_clf = xgb.XGBClassifier(random_state = 42)
lgb_clf = lgb.LGBMClassifier(random_state = 42, verbose = -1)
dt_clf = DecisionTreeClassifier(random_state = 42)
xt_clf = ExtraTreeClassifier(random_state = 42)
lsv_clf = LinearSVC(random_state = 42)
svc_clf = SVC(random_state = 42, probability = True)
rf_clf = RandomForestClassifier(random_state = 42)

x = time.time()
xgb_clf.fit(x_train, y_train)
print('XGBoost done', time.time() - x)

x = time.time()
lgb_clf.fit(x_train, y_train)
print('Light GBM done', time.time() - x)

x = time.time()
dt_clf.fit(x_train, y_train)
print('Decision Tree done', time.time() - x)

x = time.time()
xt_clf.fit(x_train, y_train)
print('Extra Tree done', time.time() - x)

x = time.time()
lsv_clf.fit(x_train, y_train)
print('Linear SVC done', time.time() - x)

x = time.time()
svc_clf.fit(x_train, y_train)
print('SVC done', time.time() - x)

x = time.time()
rf_clf.fit(x_train, y_train)
print('RandomForest done', time.time() - x)
````

````python
from sklearn.metrics import accuracy_score

pred1 = xgb_clf.predict(x_test)
pred2 = lgb_clf.predict(x_test)
pred3 = dt_clf.predict(x_test)
pred4 = xt_clf.predict(x_test)
pred5 = lsv_clf.predict(x_test)
pred6 = svc_clf.predict(x_test)
pred7 = rf_clf.predict(x_test)

acc1 = accuracy_score(y_test, pred1)
acc2 = accuracy_score(y_test, pred2)
acc3 = accuracy_score(y_test, pred3)
acc4 = accuracy_score(y_test, pred4)
acc5 = accuracy_score(y_test, pred5)
acc6 = accuracy_score(y_test, pred6)
acc7 = accuracy_score(y_test, pred7)

print(acc1)
print(acc2)
print(acc3)
print(acc4)
print(acc5)
print(acc6)
print(acc7)
````

### 이진 분류

````python
# 이미지가 5를 나타내는지 만을 판단하는 모델을 훈련시켜본다. 가장 성능이 좋았던 건 SVC지만 시간이 오래 걸리므로 LightGBM 을 사용한다.

y_train_5 = (y_train == 5)
y_test_5 = (y_test == 5)

lgb_clf = lgb.LGBMClassifier(random_state = 42, verbose = -1)
lgb_clf.fit(x_train, y_train_5)
````

````python
# 오차 행렬
from sklearn.metrics import confusion_matrix

y_train_pred = lgb_clf.predict(x_train)

confusion_matrix(y_train_5, y_train_pred)
````

오차 행렬은 다음 구조를 가지고 있다.

```
TN | FP
---|---
FN | TP
```

*   TN : True Negetive, 실제 거짓 값을 거짓으로 예측, 여기서는 5가 아닌 것을 5가 아니라고 예측
*   FP : False Positive, 실제 거짓 값을 참으로 예측, 여기서는 5가 아닌 것을 5라고 예측
*   FN : False Negetive, 실제 참 값을 거짓으로 예측, 여기서는 5인 것을 5가 아니라고 예측
*   TP : True Positive, 실제 참 값을 참으로 예측, 여기서는 5인 것을 5라고 예측



````python
# 정확도, 전체 데이터 중 맞춘 비율
# (TN + TP) / (TN + FP + FN + TP)

accuracy_score(y_train_5, y_train_pred)
````

````python
# 정밀도와 재현율
# 정밀도 : 참이라고 한 예측이 올바른 정도 TP / (FP + TP)
# 재현율 : 실제 참인 데이터를 참이라고 예측한 정도 TP / (FN + TP)
from sklearn.metrics import precision_score, recall_score

print(precision_score(y_train_5, y_train_pred))
print(recall_score(y_train_5, y_train_pred))
````

정밀도(precision)와 재현율(recall)은 한 쪽이 증가하면 한 쪽은 감소하는 **트레이드 오프** 관계이다.

````python
from sklearn.model_selection import cross_val_predict

y_proba = cross_val_predict(lgb_clf, x_train, y_train_5, cv=3,
                                    method="predict_proba")

y_proba = y_proba[:, 1]
````

````python
from sklearn.metrics import precision_recall_curve

precisions, recalls, thresholds = precision_recall_curve(y_train_5, y_proba)
````

````python
plt.figure(figsize=(8, 4))

plt.plot(thresholds, precisions[:-1], "b--", label="Precision", linewidth=2)
plt.plot(thresholds, recalls[:-1], "g-", label="Recall", linewidth=2)

threshold = 0.5
plt.vlines(threshold, 0, 1.0, "k", "dotted", label="threshold")

idx = (thresholds >= threshold).argmax()
plt.plot(thresholds[idx], precisions[idx], "bo")
plt.plot(thresholds[idx], recalls[idx], "go")

plt.axis([0, 1, 0.8, 1])
plt.grid()
plt.xlabel("Threshold")
plt.legend(loc="center right")

plt.show()
````

````python
import matplotlib.patches as patches

plt.figure(figsize=(6, 5))

plt.plot(recalls, precisions, linewidth=2, label="Precision/Recall curve")

plt.plot([recalls[idx], recalls[idx]], [0., precisions[idx]], "k:")
plt.plot([0.0, recalls[idx]], [precisions[idx], precisions[idx]], "k:")
plt.plot([recalls[idx]], [precisions[idx]], "ko",
         label="Point at threshold 0.5")

plt.gca().add_patch(patches.FancyArrowPatch(
    (0.94, 0.74), (0.85, 0.92),
    connectionstyle="arc3,rad=.2",
    arrowstyle="Simple, tail_width=1.5, head_width=8, head_length=10",
    color="#444444"))

plt.text(0.75, 0.8, "Higher\nthreshold", color="#333333")

plt.xlabel("Recall")
plt.ylabel("Precision")
plt.axis([0, 1, 0, 1])
plt.grid()
plt.legend(loc="lower left")

plt.show()
````

````python
from sklearn.metrics import roc_curve

fpr, tpr, thresholds = roc_curve(y_train_5, y_proba)

plt.figure(figsize=(6, 5))

plt.plot(fpr, tpr, '--', linewidth=2, label="ROC curve")
plt.plot([0, 1], [0, 1], 'k:', label="Random classifier's ROC curve")

plt.gca().add_patch(patches.FancyArrowPatch(
    (0.17, 0.97), (0.04, 0.78),
    connectionstyle="arc3,rad=.4",
    arrowstyle="Simple, tail_width=1.5, head_width=8, head_length=10",
    color="#444444"))
plt.text(0.09, 0.82, "Higher\nthreshold", color="#333333")
plt.xlabel('False Positive Rate (Fall-Out)')
plt.ylabel('True Positive Rate (Recall)')
plt.grid()
plt.axis([0, 1, 0, 1])
plt.legend(loc="lower right", fontsize=13)

plt.show()
````

````python
from sklearn.metrics import roc_auc_score

roc_auc_score(y_train_5, y_proba)
````

### 다중 클래스 분류

````python
# soft voting 약 25분 소요 (SVC의 5겹 교차 검증)

from sklearn.ensemble import VotingClassifier

vot_clf1 = VotingClassifier(estimators = [('xgb', xgb.XGBClassifier(random_state = 42)),
                                         ('lgb', lgb.LGBMClassifier(random_state = 42, verbose = -1)),
                                         ('svc', SVC(random_state = 42)),
                                         ('rf', RandomForestClassifier(random_state = 42))],
                           weights = [0.9795, 0.9774, 0.9792, 0.9704])

vot_clf2 = VotingClassifier(estimators = [('xgb', xgb.XGBClassifier(random_state = 42)),
                                         ('lgb', lgb.LGBMClassifier(random_state = 42, verbose = -1)),
                                         ('svc', SVC(random_state = 42, probability = True)),
                                         ('rf', RandomForestClassifier(random_state = 42))],
                           weights = [0.9795, 0.9774, 0.9792, 0.9704],
                           voting = 'soft')

x = time.time()
vot_clf1.fit(x_train, y_train)
print('hard voting done', time.time() - x)

x = time.time()
vot_clf2.fit(x_train, y_train)
print('soft voting done', time.time() - x)

vot_pred1 = vot_clf1.predict(x_test)
vot_pred2 = vot_clf2.predict(x_test)

vot_acc1 = accuracy_score(y_test, vot_pred1)
vot_acc2 = accuracy_score(y_test, vot_pred2)

print(vot_acc1)
print(vot_acc2)
````

````python
print(confusion_matrix(y_test, vot_pred1))
print()
print(confusion_matrix(y_test, vot_pred2))
````

````python
from sklearn.metrics import ConfusionMatrixDisplay

fig, axs = plt.subplots(1, 2, figsize = (10, 4))
axs = axs.flatten()

plt.rc('font', size=9)

ConfusionMatrixDisplay.from_predictions(y_test, vot_pred1, ax = axs[0])
axs[0].set_title('Hard Voting')

# ConfusionMatrixDisplay.from_predictions(y_test, vot_pred1, ax = axs[1], normalize = 'true', values_format = '.0%')
ConfusionMatrixDisplay.from_predictions(y_test, vot_pred2, ax = axs[1])
axs[1].set_title('Soft Voting')

plt.show()
````

### 데이터 증식

````python
from scipy.ndimage import shift

def image_shift(image, dx, dy):
    image = image.reshape((28, 28))
    shifted_image = shift(image, (dy, dx))
    return shifted_image
````

````python
fig, axs = plt.subplots(2, 5, figsize = (5, 2))
axs = axs.flatten()

for i in range(5):
    img = image_shift(x_train[i], 5, 0)
    plt.sca(axs[i])
    plt.imshow(x_train[i], cmap = 'binary')
    plt.axis('off')

    plt.sca(axs[i + 5])
    plt.imshow(img, cmap = 'binary')
    plt.axis('off')

plt.show()
````

````python
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()

x_train_augmented = [i for i in x_train]
y_train_augmented = [i for i in y_train]

for dx, dy in ((-2, 0), (2, 0), (0, 2), (0, -2)):
    for image, label in zip(x_train, y_train):
        x_train_augmented.append(image_shift(image, dx, dy))
        y_train_augmented.append(label)

x_train_augmented = np.array(x_train_augmented)
y_train_augmented = np.array(y_train_augmented)
````

````python
x_train_augmented = x_train_augmented / 255
x_test = x_test / 255

x_train_augmented = x_train_augmented.reshape(x_train_augmented.shape[0], -1)
x_test = x_test.reshape(x_test.shape[0], -1)
````

````python
# 노트북 CPU로 100분 소요

xgb_clf = xgb.XGBClassifier(random_state = 42)
lgb_clf = lgb.LGBMClassifier(random_state = 42, verbose = -1)
svc_clf = SVC(random_state = 42)
rf_clf = RandomForestClassifier(random_state = 42)

xgb_clf.fit(x_train_augmented, y_train_augmented)
lgb_clf.fit(x_train_augmented, y_train_augmented)
svc_clf.fit(x_train_augmented, y_train_augmented)
rf_clf.fit(x_train_augmented, y_train_augmented)

pred1 = xgb_clf.predict(x_test)
pred2 = lgb_clf.predict(x_test)
pred3 = svc_clf.predict(x_test)
pred4 = rf_clf.predict(x_test)

print(accuracy_score(y_test, pred1)) # 0.9795
print(accuracy_score(y_test, pred2)) # 0.9774
print(accuracy_score(y_test, pred3)) # 0.9792
print(accuracy_score(y_test, pred4)) # 0.9704
````

````python
# soft voting은 오래 걸리는 관계로 hard voting
# 이 코드도 Google Colab에서 4시간 소요..

vot_clf = VotingClassifier(estimators = [('xgb', xgb.XGBClassifier(random_state = 42)),
                                         ('lgb', lgb.LGBMClassifier(random_state = 42, verbose = -1)),
                                         ('svc', SVC(random_state = 42)),
                                         ('rf', RandomForestClassifier(random_state = 42))],
                           weights = [0.9813, 0.9766, 0.9874, 0.9762])

vot_clf.fit(x_train_augmented, y_train_augmented)
vot_pred = vot_clf.predict(x_test)

print(accuracy_score(y_test, vot_pred)) # 0.9797 -> 0.05 증가. 하지만 SVC 단일 모델보다 성능이 좋지 않다.
````
